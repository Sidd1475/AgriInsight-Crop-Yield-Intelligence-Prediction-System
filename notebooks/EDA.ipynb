## Summary of Exploratory Data Analysis (EDA)

1. Yield Trend Over Time
The distribution of Yield across years shows an increasing trend during the initial years, followed by a relatively stable phase, and a noticeable decline after 2012. This indicates that Year is an important temporal feature and should be considered carefully during modeling.

2. Crop Distribution
No single crop overwhelmingly dominates the dataset. This balanced distribution suggests that a generalized model across crops is feasible, provided crop is properly encoded as a categorical feature.

3. Seasonal Representation
The dataset contains mixed seasonal categories such as Whole Year, Kharif, and Rabi. These categories require further cleaning and standardization to avoid ambiguity. However, no single season significantly dominates the dataset.

4. Production Units
Production is predominantly measured in tonnes, while a smaller proportion of data is recorded in other units. This imbalance may introduce bias, and appropriate handling of unit-based variation is necessary before modeling.

5. Correlation Analysis
Yield shows a strong positive correlation with Production, while correlations with other features appear comparatively weaker. This relationship must be examined carefully to avoid potential multicollinearity issues.

6. Modeling Considerations

Log transformation may be required for skewed numerical features, including potentially the target variable (Yield).

Seasonal categories need to be cleaned and standardized.

Year emerges as a significant predictive feature.

Crop must be encoded appropriately to capture crop-specific yield variations.
